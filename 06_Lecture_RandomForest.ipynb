Data processing

```python
# load data
train_df = pd.read_csv('/Users/hela/Code/fast_ai/titanic/train.csv')
test_df = pd.read_csv('/Users/hela/Code/fast_ai/titanic/test.csv')
```

```python
# create modes
modes = train_df.mode().iloc[0]
```

```python
# preprocessing
def proc_data(df):
    df['Fare'] = df.Fare.fillna(0)
    df.fillna(modes, inplace=True)
    df['LogFare'] = np.log1p(df['Fare'])
    df['Embarked'] = pd.Categorical(df.Embarked)  # instead of dummy vs; it will look like words, but they actually will be numbers 1/0
    df['Sex'] = pd.Categorical(df.Sex)  # instead of dummy vs; it will look like words, but they actually will be numbers 1/0

proc_data(train_df)
proc_data(test_df)

# check
print(train_df.Sex.head())  # it will look like words...
print(train_df.Sex.cat.codes.head())  # ...but they actually will be numbers 1/0
```

Desicion tree  
For Random Forest/Gradient Boosting Machine, we first need a decision tree.  
For a desicion tree, we fisrt need a binary split.

A binary split  
Is when all rows are placed into one of two groups, based on whether they're above or below some threshold of some column.   
E.g.: split the rows into males and females, by using the threshold 0.5 and the column Sex.
